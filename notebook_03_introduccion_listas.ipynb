## 3.1 Creación y Acceso a Listas

```python
# Una lista es una colección ordenada y mutable
# Se define con corchetes []
lenguajes = ['Python', 'Java', 'C++', 'JavaScript', 'Go']
print(lenguajes)
print(type(lenguajes))

# Las listas pueden mezclar tipos (aunque no es recomendable)
mezcla = [42, "hola", 3.14, True, None]
print(mezcla)
```

```python
# Acceso por índice — empieza en 0 (como en la mayoría de lenguajes)
frutas = ['manzana', 'banana', 'cereza', 'durazno', 'fresa']

print(frutas[0])   # Primera: manzana
print(frutas[2])   # Tercera: cereza
print(frutas[-1])  # Última: fresa (índice negativo!)
print(frutas[-2])  # Penúltima: durazno

# Slicing — obtener sublistas [inicio:fin:paso]
print(frutas[1:3])   # ['banana', 'cereza'] — índice 1 y 2 (no incluye 3)
print(frutas[:3])    # Primeros 3 elementos
print(frutas[2:])    # Desde índice 2 hasta el final
print(frutas[::2])   # Cada 2 elementos
print(frutas[::-1])  # Lista al revés!
```

## 3.2 Modificar Listas

```python
# Las listas son MUTABLES — puedes cambiarlas después de crearlas
# (A diferencia de los strings que son inmutables)

colores = ['rojo', 'verde', 'azul']
print(f"Original: {colores}")

# Modificar elemento existente
colores[1] = 'amarillo'
print(f"Modificado: {colores}")

# Agregar al final
colores.append('morado')
print(f"Con append: {colores}")

# Insertar en posición específica
colores.insert(1, 'naranja')
print(f"Con insert: {colores}")
```

```python
# Eliminar elementos
numeros = [1, 2, 3, 4, 5, 3, 6]

# del — elimina por índice
del numeros[0]
print(f"Después de del: {numeros}")

# .remove() — elimina la primera ocurrencia de un valor
numeros.remove(3) #primer numero indicado que aparece, no la posicion
print(f"Después de remove(3): {numeros}")

# .pop() — elimina y retorna el último elemento (o el de índice dado)
ultimo = numeros.pop()
print(f"Elemento eliminado: {ultimo}")
print(f"Lista restante: {numeros}")

segundo = numeros.pop(1)
print(f"Elemento en índice 1 eliminado: {segundo}")
print(f"Lista final: {numeros}")
```

## 3.3 Organizar Listas

```python
# .sort() — ordena la lista EN SITIO (modifica la original)
autos = ['bmw', 'audi', 'toyota', 'ford', 'honda']
print(f"Original: {autos}")

autos.sort()
print(f"Ordenado A-Z: {autos}")

autos.sort(reverse=True)
print(f"Ordenado Z-A: {autos}")

# sorted() — retorna NUEVA lista ordenada (no modifica la original)
frutas = ['cereza', 'manzana', 'banana']
frutas_ordenadas = sorted(frutas)
print(f"
Frutas original: {frutas}")
print(f"Frutas ordenadas: {frutas_ordenadas}")
```

```python
# Información sobre la lista
numeros = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

print(f"Lista: {numeros}")
print(f"Longitud: {len(numeros)}")
print(f"Mínimo: {min(numeros)}")
print(f"Máximo: {max(numeros)}")
print(f"Suma: {sum(numeros)}")
print(f"¿Está el 9? {9 in numeros}")
print(f"¿Cuántos 5 hay? {numeros.count(5)}")
print(f"Índice del 9: {numeros.index(9)}")
```

### Ejercicio 3.2: Gestión de Lista de Tareas

```python
# Ejercicio 3.2 — Sistema de tareas simple
tareas = ['Revisar email', 'Escribir reporte', 'Reunión 3pm',
          'Almorzar', 'Revisar código']

print("=== LISTA DE TAREAS ===")
for i, tarea in enumerate(tareas, 1):
    print(f"{i}. {tarea}")

# Ordenar alfabéticamente
tareas_ordenadas = sorted(tareas)
print("
=== TAREAS ORDENADAS ===")
for i, tarea in enumerate(tareas_ordenadas, 1):
    print(f"{i}. {tarea}")

# Completar primera tarea
completada = tareas.pop(0)
print(f"
✅ Tarea completada: {completada}")
print(f"Tareas restantes: {len(tareas)}")
```

## 🚀 Mini Proyecto: Administrador de Lista de Compras

```python
# MINI PROYECTO: Lista de Compras
lista_compras = []

# Agregar productos
productos_iniciales = [
    ('Leche', 2),
    ('Pan', 1),
    ('Huevos', 12),
    ('Manzanas', 6),
    ('Yogur', 4)
]

# Construir la lista
for producto, cantidad in productos_iniciales:
    lista_compras.append(f"{cantidad}x {producto}")

print("=== LISTA DE COMPRAS ===")
for item in sorted(lista_compras):
    print(f"  ☐ {item}")

# Simulamos que ya compramos algunos artículos
comprados = ['2x Leche', '1x Pan']
pendientes = [item for item in lista_compras if item not in comprados]

print(f"
=== PENDIENTES ({len(pendientes)}) ===")
for item in pendientes:
    print(f"  ☐ {item}")

print(f"
=== COMPRADOS ({len(comprados)}) ===")
for item in comprados:
    print(f"  ✅ {item}")
```
