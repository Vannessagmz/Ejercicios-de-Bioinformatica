## 2.1 Variables — Tipado Dinámico

A diferencia de Java/C++, Python usa **tipado dinámico**: el tipo se asigna en tiempo de ejecución.

```python
# En Java necesitarías: String mensaje = "Hola";
# En Python:
mensaje = "Hola Python"
print(mensaje)

# Puedes reasignar con otro tipo (tipado dinámico)
x = 42        # x es int
print(type(x))
x = 3.14      # ahora x es float
print(type(x))
x = "texto"   # ahora x es str
print(type(x))
```

```python
# Reglas para nombrar variables en Python
nombre_usuario = "Ana"       # snake_case (convención Python)
nombreUsuario = "Ana"        # camelCase (válido pero no convencional)
# NombreUsuario = "Ana"      # PascalCase (se usa para clases)

# Variables pueden tener letras, números y guiones bajos
variable_1 = "válida"
# 1_variable = "inválida"   # NO puede empezar con número

# Python distingue mayúsculas y minúsculas
nombre = "Python"
Nombre = "Java"
print(nombre, Nombre)  # Son variables diferentes
```

## 2.2 Strings (Cadenas de Texto)

Los strings son secuencias de caracteres. En Python son inmutables.

```python
# Definición con comillas simples o dobles (equivalentes)
frase1 = 'Hola mundo'
frase2 = "Hola mundo"
print(frase1 == frase2)  # True — son iguales

# Útil cuando el string contiene comillas
apostrofe = "It's a great day"
cita = 'Ella dijo: "Hola"'
print(apostrofe)
print(cita)
```

```python
# Métodos de strings — Python tiene muchos métodos útiles
nombre = "  ana patricia  "

print(nombre.title())    # Cada palabra en mayúscula
print(nombre.upper())    # Todo mayúsculas
print(nombre.lower())    # Todo minúsculas
print(nombre.strip())    # Eliminar espacios al inicio y final
print(nombre.lstrip())   # Solo espacios izquierda
print(nombre.rstrip())   # Solo espacios derecha
```

```python
# f-strings — La forma moderna de insertar variables en strings
# Disponible desde Python 3.6+ (similar a template literals en JS)
nombre = "María"
edad = 25
ciudad = "Madrid"

# f-string (recomendado)
presentacion = f"Me llamo {nombre}, tengo {edad} años y vivo en {ciudad}."
print(presentacion)

# Puedes hacer operaciones dentro del f-string
print(f"En 10 años tendré {edad + 10} años")
print(f"Mi nombre en mayúsculas: {nombre.upper()}")
```

```python
# Escape characters y strings multilínea
#
 = nueva línea, 	 = tabulación
print("Línea 1
Línea 2
Línea 3")
print("Col1	Col2	Col3")

# String multilínea con triple comilla
poema = """
Las rosas son rojas,
las violetas azules,
Python es genial,
¡y tú también!
"""
print(poema)
```

## 2.3 Números: Integers y Floats

```python
# Enteros (int) — sin límite de tamaño en Python!
a = 1_000_000   # Puedes usar _ como separador de miles
b = 0xFF        # Hexadecimal
c = 0b1010      # Binario

print(a)
print(b)  # Imprime en decimal
print(c)  # Imprime en decimal

# Operaciones
print(f"Suma: {a + 1}")
print(f"División entera: {10 // 3}")
print(f"Módulo: {10 % 3}")
print(f"Potencia: {2 ** 10}")
```

```python
# Flotantes (float)
precio = 9.99
tasa = 0.18
total = precio * (1 + tasa)

print(f"Precio: ${precio}")
print(f"IVA: ${precio * tasa:.2f}")
print(f"Total: ${total:.2f}")

# Cuidado con la precisión flotante (común en todos los lenguajes)
print(0.1 + 0.2)          # No es exactamente 0.3!
print(round(0.1 + 0.2, 1))  # Solución: redondear
```

```python
# Conversión entre tipos (casting)
numero_str = "42"
numero_int = int(numero_str)
numero_float = float(numero_str)

print(type(numero_str))   # str
print(type(numero_int))   # int
print(type(numero_float)) # float

# De número a string
valor = 3.14
texto = str(valor)
print(f"El número {valor} como texto es: '{texto}' (tipo: {type(texto).__name__})")
```

### Ejercicio 2.2: Factura Simple

```python
# Ejercicio 2.2 — Crea una factura simple
producto = "Laptop"
precio_unitario = 15999.99
cantidad = 2
descuento = 0.10  # 10%
iva = 0.16        # 16%

# Calcula subtotal, descuento, iva y total
subtotal = precio_unitario * cantidad
monto_descuento = subtotal * descuento
precio_con_descuento = subtotal - monto_descuento
monto_iva = precio_con_descuento * iva
total = precio_con_descuento + monto_iva

# Imprime la factura con formato
print("=" * 45)
print(f"{'FACTURA':^45}")
print("=" * 45)
print(f"Producto: {producto}")
print(f"Precio unitario:  ${precio_unitario:>12,.2f}")
print(f"Cantidad:         {cantidad:>12}")
print(f"Subtotal:         ${subtotal:>12,.2f}")
print(f"Descuento (10%):  ${monto_descuento:>12,.2f}")
print(f"IVA (16%):        ${monto_iva:>12,.2f}")
print("-" * 45)
print(f"TOTAL:            ${total:>12,.2f}")
print("=" * 45)
```

## 🚀 Mini Proyecto: Conversor de Unidades

Crea un programa que convierta entre diferentes unidades de medida.

```python
# MINI PROYECTO: Conversor de Unidades
print("=== CONVERSOR DE UNIDADES ===
")

# Conversiones de temperatura
celsius = 100
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print("TEMPERATURA")
print(f"  {celsius}°C = {fahrenheit}°F = {kelvin}K")

# Conversiones de distancia
kilometros = 42.195  # Maratón
metros = kilometros * 1000
millas = kilometros * 0.621371
pies = metros * 3.28084

print("
DISTANCIA (Maratón)")
print(f"  {kilometros} km = {metros:.0f} m = {millas:.2f} millas = {pies:.0f} pies")

# Conversiones de velocidad
kmh = 120
mph = kmh * 0.621371
ms = kmh / 3.6

print("
VELOCIDAD")
print(f"  {kmh} km/h = {mph:.1f} mph = {ms:.1f} m/s")
```
